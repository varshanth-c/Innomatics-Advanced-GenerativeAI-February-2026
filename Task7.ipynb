```python
"""1. Smart Light Controller Program"""
class SmartLamp:
    def __init__(self, title):
        self.title = title
        self.condition = "OFF"

    def switch_on(self):
        self.condition = "ON"

    def show_state(self):
        print(self.title + " is " + self.condition)


lamp1 = SmartLamp("Bedroom Light")
lamp1.switch_on()
lamp1.show_state()
```

```python
"""2. Employee ID Card Generator"""
class StaffCard:
    def __init__(self, person, code, section):
        self.person = person
        self.code = code
        self.section = section

    def display_card(self):
        print("Employee ID Card")
        print("-----------------")
        print("Name:", self.person)

        print("ID:", self.code)
        print("Department:", self.section)


emp1 = StaffCard("Rahul", "EMP102", "IT")
emp1.display_card()
```

```python
"""3. Contact Information Saver"""
class PhoneEntry:
    def __init__(self, person_name, mobile_no):
        self.person_name = person_name

        self.mobile_no = mobile_no

    def show_contact(self):
        print("Contact Saved")
        print("Name:", self.person_name)

        print("Phone:", self.mobile_no)


entry1 = PhoneEntry("Anita", 9876543210)
entry1.show_contact()
```

```python
"""4. Product Price Label Creator"""
class ItemTag:
    def __init__(self, item_title, cost_value):
        self.item_title = item_title

        self.cost_value = cost_value

    def print_tag(self):
        print("Product:", self.item_title)

        print("Price: ₹" + str(self.cost_value))


product1 = ItemTag("Headphones", 2499)
product1.print_tag()
```

```python
"""5. Movie Review Display"""
class FilmInfo:
    def __init__(self, film_title, score):
        self.film_title = film_title

        self.score = score

    def show_review(self):
        print("Movie:", self.film_title)

        print("Rating:", self.score, "/ 5")


movie1 = FilmInfo("Inception", 4.8)
movie1.show_review()
```

```python
"""6. Parcel Delivery Details System"""
class ParcelInfo:
    def __init__(self, client, location):
        self.client = client
        self.location = location

    def print_details(self):
        print("Delivery Details")

        print("Customer:", self.client)
        print("Address:", self.location)


order1 = ParcelInfo("Suman", "Hyderabad")
order1.print_details()
```
